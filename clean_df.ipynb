```python
import pandas as pd


df = pd.read_excel('data/Матрица продаж 09042024.xlsx')
clean_df = df[~((df.loc[:, 'Таргет'] == 0))]

clean_df['Продано нежилых, м2'] = clean_df['Продано нежилых, м2'].apply(pd.to_numeric)

missing_cols, missing_rows = (
    (clean_df.isnull().sum(x) | clean_df.eq('').sum(x))
    .loc[lambda x: x.gt(0)].index
    for x in (0, 1)
)

clean_df = clean_df.drop(missing_cols, axis=1)

new_col = clean_df['Проект'].astype(str) + ' ' + clean_df['Корпус'].astype(str)
clean_df = clean_df.drop(['Проект', 'Корпус', 'Продано квартир, шт', 'Продано квартир, м2'], axis=1)
clean_df.insert(0, 'ЖК', new_col)

clean_df 
```
