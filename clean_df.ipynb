```python
import pandas as pd
import numpy as np


df = pd.read_excel('data/Матрица продаж 09042024.xlsx')
clean_df = df[~((df.loc[:, 'Таргет'] == 0))]

clean_df['Продано нежилых, м2'] = clean_df['Продано нежилых, м2'].apply(pd.to_numeric)

missing_cols, missing_rows = (
    (clean_df.isnull().sum(x) | clean_df.eq('').sum(x))
    .loc[lambda x: x.gt(0)].index
    for x in (0, 1)
)

clean_df = clean_df.drop(missing_cols, axis=1)

new_col = clean_df['Проект'].astype(str) + ' ' + clean_df['Корпус'].astype(str)
clean_df = clean_df.drop(['Проект', 'Корпус', 'Продано квартир, шт', 'Продано квартир, м2'], axis=1)
clean_df.insert(0, 'ЖК', new_col)

clean_df.head()
```

```python
df_houses = pd.read_excel('data/Рабочая_таблица_Данные_о_ЖК_Владивостока_v4_1_с_новым_атрибутом 09042024.xlsx')

df_houses.set_index(keys=['Unnamed: 0'], inplace=True)
df_houses = df_houses.T
df_houses.head()

start_col_index = df_houses.columns.get_loc('Общая сумма полученных денег с проданных квартир, руб') 
end_col_index = df_houses.columns.get_loc('Количество проданных нежилых помещений')
df_houses = df_houses.drop(df_houses.columns[start_col_index:end_col_index + 1], axis=1)
df_houses = df_houses.drop(["Вид из окна на море"], axis=1)

df_houses.head()
```

```python
result_df = clean_df.merge(df_houses, on='ЖК', how='left')
#result_df.to_excel(r'/home/doodhust/PycharmProjects/counting_cost_flates/data/result_noempty_df.xlsx', sheet_name='result_noempty_df')
```
