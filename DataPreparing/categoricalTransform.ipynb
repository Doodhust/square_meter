```python
result['Район'] = result['Район'].str.strip()
result['Основной материал фасада'] = result['Основной материал фасада'].str.strip()
result['Вспомогательный материал фасада'] = result['Вспомогательный материал фасада'].str.strip()
result['Тип дома'] = result['Тип дома'].str.strip()
result['Тип фасада'] = result['Тип фасада'].str.strip()
result['Месяц'] = result['Месяц'].str.strip()


attributes = ['Район',
              'Основной материал фасада',
             'Вспомогательный материал фасада',
             'Тип дома',
             'Тип фасада',
             'Месяц'] 
for atr in attributes:
    encoded = encoder.fit_transform(result[[atr]])
    result = pd.concat([result.drop(columns=[atr]), pd.DataFrame(encoded, columns=encoder.get_feature_names_out([atr]))], axis=1)
    
```
