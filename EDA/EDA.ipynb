```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from scipy. stats import pearsonr
import numpy as np
import matplotlib.pyplot as plt
```

```python
# result = pd.read_excel(r'/home/maxleb/Desktop/FEIP/square_meter/data/result_prepared_df.xlsx', index_col=0)
result = pd.read_excel(r'../data/result_prepared_df.xlsx', index_col=0)
result.head()
```

```python
data = result.drop(result.columns[:6], axis=1)
data = data.drop(['Таргет', 'Дата получения РС', 'Дата публикации ИНС', 'Дата сбора информации'], axis=1)
attributes = data.columns
attributes
```

```python
important = []
unimportant = []
for atr in attributes:
    p_value = []
    target = 'Таргет'

    # plt.figure(figsize=(10, 6))

    # plt.subplot(2, 2, 1)
    # plt.hist(result[atr], bins=10, color='blue', alpha=0.7)
    # plt.xlabel(atr)
    # plt.ylabel('Частота')
    # plt.title('Распределение : ' + atr)

    # plt.subplot(2, 2, 2)
    # plt.scatter(result[target], result[atr], color='green')
    # plt.xlabel(target)
    # plt.ylabel(atr)
    # plt.title('Диаграмма рассеяния')

    # plt.subplot(2, 2, 3)
    # plt.boxplot(result[atr])
    # plt.ylabel('Количество')
    # plt.title('Анализ выбросов : ' + atr)

    # plt.tight_layout()
    # plt.show()

    print(f'Корреляция и p_value: {atr}')
    p_value = pearsonr(result['Таргет'], result[atr])
    print("p_value:", p_value)
    if abs(p_value[0]) > 0.01:
        print("Корреляция ЯВЛЯЕТСЯ статистически значимой")
        important.append([atr, p_value[0]])
    else:
        print("Корреляция НЕ является статистически значимой")
        unimportant.append([atr, p_value[0]])



```

```python
important.sort(reverse=True, key=lambda x: x[1])
important
np.save('important.npy', important)
important
```
